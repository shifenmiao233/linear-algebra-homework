```python
import numpy as np
M = np.array([[0.95, 0.03],
              [0.05, 0.97]])
def solve_13(x0, label):
    print(f"\nPart ({label})")
    x = x0
    for k in range(1, 21):
        x = M @ x
        if k in [1, 5, 10, 20]: # 打印关键年份查看变化趋势
            print(f"Year {k}: City={int(x[0])}, Suburbs={int(x[1])}")
    return x
# (a) 初始人口: City 600,000, Suburbs 400,000
solve_13(np.array([600000, 400000]), "a")
# (b) 初始人口: City 350,000, Suburbs 650,000
solve_13(np.array([350000, 650000]), "b")
print(f"\n现象：在 Part (a) 中，城市人口从 600,000 开始，呈下降趋势；在 Part (b) 中，城市人口从 350,000 开始，呈上升趋势。")
print(f"结果：到第 20 年时，Part (a) 的城市人口变为 417,455，Part (b) 变为 370,282。")
print(f"规律：虽然起点（初始人口分布）完全不同，但随着时间推移两者的数值都在向同一个目标值靠拢,即最终会到达稳态。当系统达到稳态时，从城市搬到郊区的人数，正好等于从郊区搬回城市的人数。")
```
