```python
import numpy as np
from sympy import Matrix
# 49 题部分
print("第 49 题")
A47 = Matrix([
    [ 8, -3,  0, -7,  2],
    [-9,  4,  5, 11, -7],
    [ 6, -2,  2, -4,  4],
    [ 5, -1,  7,  0, 10]
])

# 由47题结果得主元索引为 (0, 1, 4)
pivots47 = (0, 1, 4)
B47 = A47[:, pivots47]

# 49题：选择一个没用过的列 v（索引 2 或 3），证明它在 B47 的 span 中
v47 = A47[:, 2]  # 选择第 3 列
x47 = B47.solve(v47)
print(f"向量 v47 成功表示为 B47 的线性组合，系数为: {x47}")

# 50 题
print("第 50 题")
A48 = Matrix([
    [12, 10, -6, -3,  7, 10],
    [-7, -6,  4,  7, -9,  5],
    [ 9,  9, -9, -5,  5, -1],
    [-4, -3,  1,  6, -8,  9],
    [ 8,  7, -5, -9, 11, -8]
])

# 根据你的运行结果，主元索引为 (0, 1, 3, 5)
pivots48 = (0, 1, 3, 5)
B48 = A48[:, pivots48]

# 50题：选择一个没用过的列 v（索引 2 或 4），证明它在 B48 的 span 中
v48 = A48[:, 2]  # 选择第 3 列
x48 = B48.solve(v48)
print(f"向量 v48 成功表示为 B48 的线性组合，系数为: {x48}")
print("通过 49 和 50 题，我们发现：")
print("矩阵 B 的列向量构成了 A 的列空间的基 (Basis)，矩阵 A 中任何不属于 B 的列向量 v，本质上都是多余的线性相关项。")
print("因此，这些向量 v 必然可以由基 B 的线性组合生成。")
```
