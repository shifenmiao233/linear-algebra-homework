```python
import numpy as np

t_data = np.array([0, 2, 4, 6, 8, 10])
f_data = np.array([0, 2.90, 14.8, 39.6, 74.3, 119])

X = np.column_stack([t_data**0, t_data**1, t_data**2, t_data**3])

coeffs, residuals, rank, s = np.linalg.lstsq(X, f_data, rcond=None)

t_target = 7.5
force_pred = coeffs[0] + coeffs[1]*t_target + coeffs[2]*(t_target**2) + coeffs[3]*(t_target**3)

print(f"拟合系数 (a0 到 a3): {coeffs}")
print(f"t=7.5时阻力: {force_pred:.2f} ")
```
