```python
import numpy as np

matrices = {
    "45": np.array([[-5, 10, -5,  4], [ 8,  3, -4,  7], [ 4, -9,  5, -3], [-3, -2,  5,  4]]),
    "46": np.array([[ 7,  5,  4, -9], [10,  6, 16, -4], [12,  8, 12,  7], [-8, -6, -2,  5]]),
    "47": np.array([[ 4, -7,  3,  7,  5], [ 6, -8,  5, 12, -8], [-7, 10, -8, -9, 14], [ 3, -5,  4,  2, -6], [-5,  6, -6, -7,  3]]),
    "48": np.array([[ 9, 13,  5,  6, -1], [14, 15, -7, -6,  4], [-8, -9, 12, -5, -9], [-5, -6, -8,  9,  8], [13, 14, 15,  2, 11]])
}
for task_id, A in matrices.items():
    m, n = A.shape
    rank = np.linalg.matrix_rank(A)
    print(f"第{task_id} 题")
    print(f"秩 (Rank): {rank}, 形状: {m}x{n}")
    if task_id in ["45", "46"]:
        # 45-46 判断单设(秩是否等于列数 n)
        is_one_to_one = (rank == n)
        print(f"是否为单射: {is_one_to_one}")
    else:
        # 47-48 判断满射(秩是否等于行数 m=5)
        is_onto = (rank == m)
        print(f"是否为满射: {is_onto}")
    print()
```
