```python
import numpy as np

# 将四道题的矩阵放入一个字典中，方便循环处理
matrices = {
    "47": np.array([[ 7,  2, -5,  8],
                    [-5, -3,  4, -9],
                    [ 6, 10, -2,  7],
                    [-7,  9,  2, 15]]),
    
    "48": np.array([[ 5, -7, -4,  9],
                    [ 6, -8, -7,  5],
                    [ 4, -4, -9, -9],
                    [-9, 11, 16,  7]]),
    
    "49": np.array([[12, -7, 11, -9,  5],
                    [-9,  4, -8,  7, -3],
                    [-6, 11, -7,  3, -9],
                    [ 4, -6, 10, -5, 12]]),
    
    "50": np.array([[ 8, 11, -6, -7, 13],
                    [-7, -8,  5,  6, -9],
                    [11,  7, -7, -9, -6],
                    [-3,  4,  1,  8,  7]])
}

print("--- 矩阵张成（Span）判定结果 ---")

for task_id, mat in matrices.items():
    # 计算秩
    rank = np.linalg.matrix_rank(mat)
    # 判断是否等于 R^4 的维度 4
    spans_R4 = (rank == 4)
    
    print(f"题号 {task_id}: 秩 = {rank} -> {'张成 R^4' if spans_R4 else '不张成 R^4'}")
```
