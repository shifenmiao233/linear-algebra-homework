```python
import numpy as np
from sympy import Matrix

# --- 第 47 题 ---
A47 = Matrix([
    [ 8, -3,  0, -7,  2],
    [-9,  4,  5, 11, -7],
    [ 6, -2,  2, -4,  4],
    [ 5, -1,  7,  0, 10]
])
# 求RREF，寻找主元列索引
rref47, pivots47 = A47.rref()
print(f"47题的主元列索引为: {pivots47}")
# 构造 B (提取这些主元列)
B47 = A47[:, pivots47]
print(f"矩阵 B47 的形状: {B47.shape}")
# 验证 Bx = 0 是否只有平凡解 (检查零空间是否为空)
null_space_47 = B47.nullspace()
print(f"47题 Bx=0 是否只有平凡解: {null_space_47 == []}")
print(f"\n")
# --- 第 48 题 ---
A48 = Matrix([
    [12, 10, -6, -3,  7, 10],
    [-7, -6,  4,  7, -9,  5],
    [ 9,  9, -9, -5,  5, -1],
    [-4, -3,  1,  6, -8,  9],
    [ 8,  7, -5, -9, 11, -8]
])
# 寻找主元列
rref48, pivots48 = A48.rref()
print(f"48题的主元列索引为: {pivots48}")
# 构造 B
B48 = A48[:, pivots48]
print(f"矩阵 B48 的形状: {B48.shape}")
# 验证
null_space_48 = B48.nullspace()
print(f"48题 Bx=0 是否只有平凡解: {null_space_48 == []}")
```
